```python
# Lagrange interpolation

# Importing libraries
from math import *
from sympy import *
from numpy.polynomial import polynomial as P
import sys

# Defining x as a symbol
x = symbols('x')

# Taking input
num_points = int(input("Enter the number of points (between 2 and 4): "))
while (num_points < 2 or num_points > 4): # Validation
    num_points = int(input("Number must be between 2 and 4 inclusive. Re-enter: "))
print("") # To leave a blank line

points = []
for i in range(0, num_points):
    print("Enter x coordinate for point {}: " .format(i+1))
    x_coord = float(input())
    print("Enter y coordinate for point {}: " .format(i+1))
    y_coord = float(input())
    for j in points: # Validation to ensure that the same point is not entered twice!
        if (x_coord == j[0]):
            print("The same point cannot be entered twice!")
            sys.exit()
    points.append((x_coord, y_coord))
    print("") # To leave a blank line
value = float(input("Enter x coordinate of the point at which you want to evaluate the function: "))

# Ensuring that value is within range
x_coords = []
for i in points:
    x_coords.append(i[0])
smallest = min(x_coords)
largest = max(x_coords)
while (value < smallest or value > largest):
    print("Value must be within {} and {} inclusive. Re-enter: " .format(smallest, largest))
    value = float(input())
print("") # To leave a blank line

# Constructing the polynomial
px = (0)
for i in range(0, num_points):
    numerator = (points[i][1])
    denominator = (1)
    for k in range(0, num_points):
        if (k == i):
            continue
        mul = (-points[k][0], 1)
        numerator = P.polymul(numerator, mul)
        mul = (points[i][0] - points[k][0])
        denominator = P.polymul(denominator, mul)

    numerator = P.polydiv(numerator, denominator)[0]
    px = P.polyadd(px, numerator)

# Approximating the value provided by the user
ans = P.polyval(value, px)

# Converting the polynomial into an output expression
expr = None
if (num_points == 2):
    expr = px[1]*x + px[0]
elif (num_points == 3):
    expr = px[2]*x**2 + px[1]*x + px[0]
else:
    expr = px[3]*x**3 + px[2]*x**2 + px[1]*x + px[0]

# Output
print("p({}) = {}" .format(value, ans))
print("p(x): ")
expr
```
